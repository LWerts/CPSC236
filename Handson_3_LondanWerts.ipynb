## 4.1 Even or Odd Checker
Create a program that checks whether a number is even or odd.

### Console:
```powershell

```

### Specifications:
- Store the code that gets user input and displays output in the main function.
- Store the code that checks whether the number is even or odd in a separate function.
- Assume that the user will enter a valid integer.


````python
### CODE HERE ###
def checkEvenOdd(number):
    """Function to check if even or odd"""
    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")

def main():
    """Function to take input and display output"""
    print("Check if your number is even or odd")
    num = int(input("Enter a number: "))
    result = checkEvenOdd(num)
    print(result)
````

`meters = feet * 0.3048`

- Store the code that performs the conversions in functions within a module. For example, store the code that converts feet to meters in a function in a module.

````python
### CODE HERE ###
def feetToMeters(feet):
    """Convert feet to meters"""
    print(round(feet * 0.3048, 2)

def metersToFeet(meters):
    """Convert meters to feet"""
    print(round(meters / 0.3048, 2)

def menu():
        """Displays conversion menu"""
        print("Conversion Menu: ")
        print("A. Feet to Meters")
        print("B. Meters to Feet")

def main():
    """Main function to run program"""
    print("Feet to Meters Converter")

    while True:
        menu()
        choice = input("Select a conversion (A or B): ")

    if choice == 'a':
        feet = float(input("Enter feet: "))
        print(f"{feetToMeters(feet)} meters")
    elif choice == 'b':
        meters = float(input("Enter meters: "))
        print(f"{metersToFeet(meters)} meters")
    else:
        print("Plese selecct A or B")
````

## 4.3 - Sales Tax Calculator
Create a program that uses a separate module to calculate sales tax and total after tax.


### Specifications
- The sales tax rate should be 6% of the total.
- Store the sales tax rate in a module. This module should also contain functions that calculate the sales tax and the total after tax. These functions should round the results to a maximum of two decimal places. Upload the seperate sales tax module to your GitHub repo when submitting.
- Store the code that gets input and displays output in this notebook. Divide this code into functions wherever you think it would make that code easier to read and maintain.
- Assume the user will enter valid data.



````python
### CODE HERE ###
# tax_module.py

SALES_TAX_RATE = 0.06  # 6% sales tax

def calculate_sales_tax(total):
    """Calculates sales tax based on the total amount."""
    return round(total * SALES_TAX_RATE, 2)

def calculate_total_after_tax(total):
    """Calculates the total amount after adding sales tax."""
    return round(total + calculate_sales_tax(total), 2)


import tax_module

def get_item_prices():
    """Gets item prices from the user until 0 is entered."""
    prices = []
    print("\nENTER ITEMS (ENTER 0 TO END)")
    while True:
        price = float(input("Cost of item: "))
        if price == 0:
            break
        prices.append(price)
    return prices

def main():
    """Main function to run the sales tax calculator."""
    print("Sales Tax Calculator\n")

    while True:
        prices = get_item_prices()
        total = round(sum(prices), 2)
        sales_tax = tax_module.calculate_sales_tax(total)
        total_after_tax = tax_module.calculate_total_after_tax(total)

        print(f"\nTotal:\t{total}")
        print(f"Sales tax:\t{sales_tax}")
        print(f"Total after tax:\t{total_after_tax}\n")

        again = input("Again? (y/n): ").strip().lower()
        if again != 'y':
            print("\nThanks, bye!")
            break
````

## 4.4 - Prime Number Checker
Create a program that checks whether a number is a prime number and displays the total number of factors if it is not a prime number.

### Console
```powershell

Please enter an integer between 1 and 5000: 1

Please enter an integer between 1 and 5000: 6

```

### Specifications
- A prime number is only divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the number is not a prime number, the program should display its number of factors. For example, 6 has four factors (1, 2, 3, and 6).
- Store the code that gets a valid integer for this program in its own function.
- Store the code that calculates the number of factors for a number in its own function.
- Store the rest of the code that gets input and displays output in the main function.



````python
### CODE HERE ###
# prime_module.py

def is_prime(number):
    """Returns True if the number is prime, else False. Also returns its total number of factors."""
    if number < 2:
        return False, 1  # 1 is not prime, and only has one factor (itself)

    factors = [i for i in range(1, number + 1) if number % i == 0]
    return len(factors) == 2, len(factors)

import prime_module

def main():
    """Main function to check if a number is prime and display its factors."""
    print("Prime Number Checker\n")

    while True:
        num = int(input("Enter an integer: "))
        is_prime, total_factors = prime_module.is_prime(num)

        if is_prime:
            print(f"{num} is a prime number.")
        else:
            print(f"{num} is not a prime number. It has {total_factors} factors.")

        again = input("\nWould you like to check another number? (y/n): ").strip().lower()
        if again != 'y':
            print("\nThanks, bye!")
            break
````
